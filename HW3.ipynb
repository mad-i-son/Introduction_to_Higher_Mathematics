Напишите код, который будет переводить полярные координаты в декартовы.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math
```

```python
def polar_coord(alfa, R):
    x = R * math.cos(math.radians(alfa))
    y = R * math.sin(math.radians(alfa))

    return x, y
```

```python
x, y = polar_coord(45, 5)
```

4. Задание (в программе)
exp(x) + x∙(1 – y) = 1  
y = x2 – 1

```python
from scipy.optimize import fsolve
from math import exp

def equations(vars):
    x, y = vars
    eq1 = x**2 - 1 - y 
    eq2 = exp(x) + x*(1-y) - 1
    return [eq1, eq2]

x, y =  fsolve(equations, (1, 1))

print(x, y)
```
