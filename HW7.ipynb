1. Решите линейную систему:

$\left(
\begin{array}{cccc}
1 & 2 & 3\\
4 & 0 & 6\\
7 & 8 & 9
\end{array}
\right) *     X = \left(
\begin{array}{cccc}
12\\2\\1
\end{array}
\right)$
## $A\cdot X = B$


```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[1, 2, 3], [4, 0, 6], [7, 8, 9]])
B = np.array([12, 2, 1])

np.linalg.solve(A, B)
```

4. Вычислите LU-разложение матрицы:

$\left(
\begin{array}{cccc}
1 & 2 & 3\\
2 & 16 & 21\\
4 & 28 & 73
\end{array}
\right)$

```python
import scipy 
import scipy.linalg  

A = np.array([ [1, 2, 3], [2, 16, 21], [4, 28, 73]])
P, L, U = scipy.linalg.lu(A)

print("A = ", A)
print("P = ", P)
print("L = ", L)
print("U = ", U)

print(np.dot(P.transpose(), A) - np.dot(L, U))
```
